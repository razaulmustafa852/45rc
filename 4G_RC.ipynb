```python
import matplotlib.pyplot as plt
#plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import numpy as np
from matplotlib.text import OffsetFrom
```

```python
df = pd.read_csv("4g_rc - 45n.csv")
```

```python
df.columns
```

```python
bba=df[(df['ExpID']==2)]
elastic=df[(df['ExpID']==1)]
conv=df[(df['ExpID']==9)]
```

```python
matplotlib.rc('axes',edgecolor='#D3D3D3')
ax = plt.figure()
plt.rcParams['axes.facecolor'] = 'white'
plt.plot(bba['timeslot sec'], bba['Single EMA'],label='BBB')
#plt.plot(bba['timeslot sec'], bba['Double EMA'],label='BBB')
#
#plt.plot(bba['timeslot sec'], bba['CUSUM'],label='BBB')
plt.grid(True)
plt.ylabel("Throughput in Bps")
plt.xlabel("Four Minutes Stream")
plt.legend(loc="upper center")
```

```python
fig,a =  plt.subplots(2,2,)
import numpy as np
a[0][0].plot(bba['timeslot sec'],bba['Throughput'], color='blue')
a[0][0].set_title('Throughput')
a[0][1].plot(bba['timeslot sec'],bba['Single EMA'],color='green')
a[0][1].set_title('EMA')
a[1][0].plot(bba['timeslot sec'],bba['Double EMA'],color='orange')
a[1][0].set_title('DEMA')
a[1][1].plot(bba['timeslot sec'],bba['CUSUM'],color='brown')
a[1][1].set_title('CUSUM')
plt.tight_layout()
plt.show()
```
